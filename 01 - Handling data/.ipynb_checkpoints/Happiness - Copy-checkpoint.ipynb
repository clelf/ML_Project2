## Data from World Happiness Report

The World Happiness Report is an annual publication of the United Nations Sustainable Development Solutions Network. It contains articles, and rankings of national happiness based on respondent ratings of their own lives, which the report also correlates with various life factors.

In this notebook we will explore the happiness of different countries and the features associated.
The datasets that we will use are available in *Data*: **happiness2020.pkl** and **countries_info.csv**.

Although the features are self-explanatory, here a summary: 

**happiness2020.pkl**
* country: *Name of the country*
* happiness_score: *Happiness score*
* social_support: *Social support (mitigation the effects of inequality)*
* healthy_life_expectancy: *Healthy Life Expectancy*
* freedom_of_choices: *Freedom to make life choices*
* generosity: *Generosity (charity, volunteers)*
* perception_of_corruption: *Corruption Perception*
* world_region: *Area of the world of the country*

**countries_info.csv**
* country_name: *Name of the country*
* area: *Area in sq mi*
* population: *Number of people*
* literacy: *Literacy percentage*

```python
import pandas as pd
import numpy as np
%matplotlib inline

DATA_FOLDER = 'Data/'

HAPPINESS_DATASET = DATA_FOLDER+"happiness2020.csv"
COUNTRIES_DATASET = DATA_FOLDER+"countries_info.csv"
```

## Task 1: Load the data

Load the 2 datasets in Pandas dataframes (called *happiness* and *countries*), and show the first rows.


**Hint**: Use the correct reader and verify the data has the expected format.

```python
df_happiness = pd.read_csv(HAPPINESS_DATASET)
df_countries = pd.read_csv(COUNTRIES_DATASET)

df_happiness.head()
```

```python
df_countries.head()
```

## Task 2: Let's merge the data

Create a dataframe called *country_features* by merging *happiness* and *countries*. A row of this dataframe must describe all the features that we have about a country.

**Hint**: Verify to have all the rows in the final dataframe

```python
# Write your code here
country_features = pd.concat([df_happiness, df_countries], axis=1)
country_features.head()
```

## Task 3: Where do people are happier?

Print the top 10 countries based on their happiness score (high is better).

```python
# Write your code here
country_features[['country', 'happiness_score']].sort_values(ascending = False, by='happiness_score').head(10)
```

We are interested to know in what world region the people are happier. 

Create and print a dataframe with the (1) average happiness score and (2) the number of contries for each world region.
Sort the result to show the happiness ranking.

```python
### New dataframe with world_region as an index
country_features_WR = country_features.set_index(['world_region', 'country']).sort_values(by='world_region')
country_features_WR.head()
```

```python
# 3 Seperate np arrays containing all neccesary features
world_regions = country_features['world_region'].unique()
average_happiness = np.zeros(len(world_regions))
n_countries = np.zeros(len(world_regions))

# Update each region with that regions average happiness and n_countries
for i, w in enumerate(average_happiness):
    average_happiness[i] = country_features_WR.loc[world_regions[i], 'happiness_score'].mean()
    n_countries[i] = country_features_WR.loc[world_regions[i], 'happiness_score'].size
```

```python
# Concatenate and sort to final dataframe
world_region_happiness = pd.DataFrame([world_regions, average_happiness, n_countries])
world_region_happiness = world_region_happiness.T
world_region_happiness.columns = ['world_region', 'average_happiness', 'n_countries']
world_region_happiness = world_region_happiness.sort_values(by='average_happiness', ascending=False)
print(world_region_happiness.to_string(index=False))
```

The first region has only a few countries! What are them and what is their score?

```python
# Write your code here
happiest_region = country_features_WR.loc['North America and ANZ', 'happiness_score']
happiest_region
```

## Task 4: How literate is the world?

Print the name of countries with a level of literacy of 100%. 

For each country, print the name and the world region with the format: *{region name} - {country name} ({happiness score})*

```python
# Slice neccesary data and add converted literacy from strings to float
literacy_all = country_features[['world_region', 'country']].copy()
literacy_all['literacy'] = country_features['literacy'].str.replace(',', '.').astype(float)

# Select 100% literacy and print
literacy_complete = literacy_all[literacy_all['literacy'] == 100]
print(literacy_complete.to_string(index=False))
```

What is the global average?

```python
# Print global average - skipping the 2 NaN values
print(literacy_all['literacy'].mean(skipna=True))
```

Calculate the proportion of countries with a literacy level below 50%. Print the value in percentage, formatted with 2 decimals.

```python
# Calculate proportions - .count() skips NaN by default
n_literacy_below50 = literacy_all.loc[literacy_all['literacy'] < 50, 'literacy'].count()
n_totall = literacy_all['literacy'].count()

# Culculate percentage and print formated number
literacy_below_50 = n_literacy_below50/n_totall*100
print("%.2f" % literacy_below_50)
```

Print the raw number and the percentage of world population that is illiterate.

```python
# Investigating 'Series' data for nans and faulty values
str_col = 'population'
df_data_investigation = pd.DataFrame([{'column_name': str_col,
                                       'type': type(country_features[str_col].array[0]),
                                       'mean': country_features[str_col].array.mean(),
                                       'range': np.array([min(country_features[str_col].array), 
                                                         max(country_features[str_col].array)]),
                                       'n_vals': country_features[str_col].shape[0],
                                       'n_NaN': np.isnan(country_features[str_col].array).sum() }])
df_data_investigation
```

```python
# Add column for illiterate population to df
pd.options.display.float_format = '{:.2f}'.format

illiteracy_all = pd.concat([literacy_all, country_features['population']], axis=1).copy()
illiteracy_all['illiterate_population'] = illiteracy_all.population.mul((100-literacy_all.literacy)/100, axis=0)

illiteracy_all.head()
```

```python
#Calulcate values and print
n_illiterate_population = illiteracy_all['illiterate_population'].sum()
proportion_illiteracy = 100*n_illiterate_population/illiteracy_all['population'].sum()
print({'Illiterate w_pop ': n_illiterate_population, 'percentage of totall': proportion_illiteracy.round()})
```

## Task 5: Population density

Add to the dataframe a new field called *population_density* computed by dividing *population* by *area*.

```python
# Write your code here
str_col = 'area'
df_data_investigation = pd.DataFrame([{'column_name': str_col,
                                       'type': type(country_features[str_col].array[0]),
                                       'mean': country_features[str_col].array.mean(),
                                       'range': np.array([min(country_features[str_col].array), 
                                                         max(country_features[str_col].array)]),
                                       'n_vals': country_features[str_col].shape[0],
                                       'n_NaN': np.isnan(country_features[str_col].array).sum() }])
df_data_investigation
```

```python
country_features['population_density'] = country_features['population']/country_features['area']
country_features.head()
```

What is the happiness score of the 3 countries with lowest population density?

```python
# Write your code here
country_features[['country', 'happiness_score', 'population_density']].nsmallest(3, 'population_density', keep='all')
```
